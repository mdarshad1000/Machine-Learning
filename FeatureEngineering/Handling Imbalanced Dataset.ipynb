#### Imbalanced Dataset
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('credit_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

Here the dataset is clearly imbalanced as 284315 vs 492

0 --> Legit transactions
1 --> Fraudulent transactions

```python
# Separating the legit and frauduelent transactions
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

##### Undersampling
To handle imbalanced data - where equal no. of datapoints are taken from both classes

```python
# Build a sample dataset containing similar distribution of Legit & Fraudulent
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
new_df.head()
```
