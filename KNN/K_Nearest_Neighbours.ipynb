* Difference distance criteria

* Choosing optimum K value
* Less efficient with high dimesnional data
* Doesn't perform well on imbalanced dataset
* Sensitive to Outliers

* Manhattan distance is preferred over Eucldean distance when there is high dimensionality in the data






```python
# Two distinct two dimesnional points
p1 = (1,1)
p2 = (2,2)
```

```python
# Calculating Euclidean distance

dist = (p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2

import numpy as np

euclidean_dist = np.sqrt(dist)
print(euclidean_dist)
```

```python
# Consider four dimensional points
p1 = (1,1,1,1)
p2 = (2,2,2,2)

dist = 0 

for i in range(len(p1)):
  dist = dist + (p1[i] - p2[2]) ** 2

euclidean_dist = np.sqrt(dist)
print(euclidean_dist)
```

Function to calculate Euclidean distance

```python
def get_euclidean_distance(x1, x2):

  dist = 0

  for item in range(len(x1)):

    dist += (x1[item] - x2[item]) ** 2

  euclidean_dist = np.sqrt(dist)
  print(euclidean_dist)
```

```python
get_euclidean_distance((1,1,1,1), (2,2,2,2))
```

Similarly, Function to calculate Manhattan function


```python
def get_manhattan_distance(x1, x2):

  dist = 0

  for item in range(len(x1)):

    dist += abs(x1[item] - x2[item])

  print(dist)
```

```python
get_manhattan_distance((1,1,1,1), (2,2,2,2))
```
